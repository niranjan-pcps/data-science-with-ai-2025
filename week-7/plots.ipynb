```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {
    "Year": [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    "Sales": [100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600]
}
df = pd.DataFrame(data)
df
```

```python
# using Matplotlib

plt.bar(df["Year"], df["Sales"],color="blue")
plt.title("Sales by Year")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

```python
# using seaborn
plt.figure(figsize=(10, 5))
sns.barplot(x='Year',y='Sales',data=df)
plt.title("Sales by Year")
```

```python
# line plot using matplotlib
plt.plot(df["Year"], df["Sales"],color="red",linestyle="-.",marker="x")
plt.title("Sales by Year")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

```python
# using seaborn for line plot
sns.lineplot(x="Year",y="Sales",data=df,linestyle="--",marker="o")
plt.title("Sales by Year")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
# histogram using matplotlib
import numpy as np
data = np.random.normal(loc=50, scale=10,size=200)
df = pd.DataFrame({'Score':data})
df
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['Score'],bins=5,color='orange',edgecolor='green')
plt.title('Distribution of Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
sns.histplot(data=df, x="Score",bins=20,kde=True,color='skyblue')
plt.title('Distribution of Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Boxplot
math_scores = [65,70,75,80,85,90,95,200]
english_scores = [55,60,65,70,75,80,85,90]

plt.figure(figsize=(10,5))
plt.boxplot([math_scores,english_scores],labels=['Math','English'])
plt.title('Boxplot of Math and English Scores')
plt.xlabel('Subject')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

```python
# using seaborn
df = pd.DataFrame({
    'subject': ["Math"]*7 +["Science"]*7,
    'score': [88,92,80,89,100,76,95,85,92,30,40,60,70,80]
})
sns.boxplot(x="subject",y="score",data=df)
plt.title('Seaborn Box Plot')
plt.show()
```

```python
# Heatmap 
data = np.array([[30,20,40],
                [25,35,30],
                [20,25,45]
])
df = pd.DataFrame(data, columns=["Math","Science","English"], 
                  index=["class_a","class_b","class_c"])
# using matlpltolib
plt.imshow(data,cmap='YlGnBu')
plt.title('Heatmap')
plt.xticks([0,1,2],df.columns.tolist())
plt.yticks([0,1,2],["class_a","class_b","class_c"])
plt.colorbar(label='scores')
plt.show()
```

```python
sns.heatmap(df,annot=True, cmap="YlGnBu")
```
