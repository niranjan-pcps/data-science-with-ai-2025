```python
import matplotlib.pyplot as plt
import numpy as np

grades = ['grade_a','grade_b','grade_c','grade_d','grade_e']
grades_values = [60,65,45,80,90]


grades_array = np.array(grades)
grades_values_array = np.array(grades_values)

plt.bar(grades_array,grades_values_array)
plt.show()
```
