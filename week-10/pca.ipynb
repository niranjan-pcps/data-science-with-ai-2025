```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
data = load_iris()
X = data.data
y = data.target
```

```python
print("original shape", X.shape)
```

```python
plt.subplot(1, 2,1)
plt.scatter(X[:,0], X[:,1],c=y, cmap='viridis',edgecolor='k',s =40)
plt.title('Before PCA for first two features')
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
plt.subplot(1,2,2)
plt.scatter(X_pca[:,0],X_pca[:,1],c=y, cmap='viridis', edgecolor='k',s=40)
plt.title('After PCA (2 principal components)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.tight_layout()
plt.show()
```

```python
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_pca[:,0],X_pca[:,1],X_pca[:,2], c=y, cmap='viridis',edgecolor='k',s=40)
ax.set_title('After PCA (3 Pricnipal COmponents)')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

plt.legend(*scatter.legend_elements(),title="classes",loc="best")
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame(X_pca, columns=['PC1','PC2','PC3'])
feature_names = data.feature_names
target_names = data.target_names
df['Species'] = [target_names[label] for label in y]
```

```python
import plotly.express as px
fig = px.scatter_3d(df, x=X_pca[:,0],y=X_pca[:,1],z = X_pca[:,2],
                    color='Species', title='Iris Dataset',
                    labels = {"PC1","PC2","PC3"})
fig.show()
```
