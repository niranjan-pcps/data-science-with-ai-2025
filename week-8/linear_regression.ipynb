```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(-1,1, 100).reshape(-1,1)
```

```python
y = 2*x + 1 + 0.4 * np.random.randn(*x.shape)
```

```python
#Intialize weights and biases
w = np.random.randn()
b = np.random.randn()
```

```python
# Hyperparameters
lr = 0.001 # learning rate
epochs = 1000 # number of epochs
```

```python
losses = []
```

```python
for epoch in range(epochs):
  y_pred = w*x + b
  loss = np.mean((y-y_pred)**2)
  losses.append(loss)

  # Gradients
  dw = np.mean(2*x*(y_pred-y))
  db = np.mean (2* (y_pred-y))

  #Update weights and biases
  w -= lr* dw
  b -= lr* db
  print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss:.4f}, w: {w:.4f}, b:{b:.4f}")
```

```python
plt.plot(range(1, epochs+1),losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training loss over epochs')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter( x,y, color='blue',label='Original Data')
plt.plot(x,w*x+b, color='red',linewidth=2, label='Learned line of fit')
plt.title('Linear Regression -Line of Fit')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
